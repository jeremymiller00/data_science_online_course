```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars = pd.read_csv('cars_multivariate.txt')
```

```python
cars.corr()
```

```python
cars.describe()
```

```python
# OO matplotlib
# create empty figure, add set of axes, add parameters
cars.head()
```

```python
cars_fig = plt.figure()
axes = cars_fig.add_axes([.1,.1,.8,.8])
axes.scatter(cars['weight'], cars['mpg'])
axes.set_xlabel('Weight')
axes.set_ylabel('MPG')
axes.set_title('Weight vs MPG')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.52,.58,.35,.25])
axes1.scatter(cars['weight'], cars['mpg'])
axes2.scatter(cars['displacement'], cars['weight'])
axes2.set_title('Disp vs. Weight')
axes1.set_title('Weight vs. MPG')
```

```python
# oop subplots
fig,axes = plt.subplots(nrows = 1, ncols = 2)
# tuple unpacking, axes manager on top of plt.figure (functional method)
# subplots add axes for you based on rows and columns
# axes.scatter(cars['weight'], cars['mpg'])
# for ax in axes:
#     ax.scatter(cars['weight'], cars['mpg'])
axes[0].scatter(cars['weight'], cars['mpg'])
axes[1].scatter(cars['displacement'], cars['weight'])
axes[0].set_title('Weight vs MPG')
axes[1].set_title('Disp vs Wieght')
axes[0].set_xlabel('Weight')
axes[0].set_ylabel('MPG')
axes[1].set_xlabel('Disp')
axes[1].set_ylabel('Weight')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(10,5))
# ax = fig.add_axes([0,0,1,1])
axes[0].scatter(cars['weight'], cars['mpg'])
axes[1].scatter(cars['displacement'], cars['weight'])
axes[0].set_title('Weight vs MPG')
axes[1].set_title('Disp vs Wieght')
axes[0].set_xlabel('Weight')
axes[0].set_ylabel('MPG')
axes[1].set_xlabel('Disp')
axes[1].set_ylabel('Weight')
plt.tight_layout()
```

```python
fig.savefig('weight_disp_mpg.png', dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1,])
axes.scatter(cars['weight'], cars['mpg'], color = 'green', label = 'Weight vs. MPG')
axes.scatter(cars['weight'], cars['cylinders'], color = 'orange', label = 'Cylinders')
plt.tight_layout()
axes.legend()
```

```python
x = np.array(sorted(np.random.normal(0,1,20)))
x
```

```python
y = x**2
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# axes.plot(x,y)
axes.plot(x,y, color='blue', lw = 2, alpha = 0.5, ls = '--', marker = 'o', ms = 20, 
          markerfacecolor = 'yellow', markeredgewidth = 3)
axes.plot(y,x, color = 'red', lw = 2, alpha = 0.5, ls = '--', marker = '>', ms = 20,
         markerfacecolor = 'green', markeredgewidth = 3)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color = 'orange', lw = 2, ls = '--')
axes.set_xlim(-1, 1)
axes.set_ylim(0,1)
```
