```python
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set_style(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x=tips['total_bill'], y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x=tips['total_bill'], y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'], y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'], height=0.1)
```

Categorical Plot Types: 
Boxplot, violinplot, stripplot, swarmplot, barplot, countplot

```python
sns.barplot(x='sex', y='tip_pct', data=tips, ci='sd')
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.boxplot(x='day', y='tip_pct', data=tips, palette='rainbow', showfliers=False)
```

```python
sns.boxplot(x='day', y='tip_pct', data=tips, hue='sex', showfliers=False)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, split=True, hue='sex')
```

```python
sns.swarmplot(x='day', y='tip_pct', data=tips, color='black', size=3)
sns.violinplot(x='day', y='tip_pct', data=tips)
```

```python
sns.factorplot(x='day', y='tip_pct', data=tips, kind='violin')
```

```python
sns.factorplot(x='smoker', y='tip_pct', data=tips, kind='bar', hue = 'sex')
```
